### 7:  Business Decisions - **3 points**

The owner of the company that oversees all the breweries needs your help to decide which beers to promote and which should be re-designed to improve its customer acceptance.

NOTE - For this, let's use a new measure called `goodness` which will depend on the value of the `review/overall` value.

|goodness|min_score|max_score|
|---|---|---|
|0_bad|0.0|3.0|
|1_neutral|3.5|4.0|
|2_great|4.5|5.0|


**7.1**: Create the column `goodness` - **1 point**

Hint: You can use conditionals like `df.loc` or `np.where` to create a column based on another column's value

```python
# YOUR CODE HERE
def goodness(review):
    if 0.0 <= review <= 3.0:
        return '0_bad'
    elif 3.5 <= review <= 4.0:
        return '1_neutral'
    elif 4.5 <= review <= 5.0:
        return '2_great'
   
beer_reviews['goodness']=beer_reviews['review/overall'].map(goodness)
beer_reviews.head(5)
```

**7.2**: Create a table called `pv_beer_reviews` with the top 20 beers (`beer/name`) and the percentage of evaluations that are bad, neutral and great (pivot table). - **1 point**

A table like the following is expected.

|goodness|0_bad|1_neutral|2_great|
|---|---|---|---|
|beer1|0.12|0.30|0.58|
|beer2|0.25|0.35|0.40|
|...|...|...|...|
|beer20|0.08|0.50|0.42|

Hint: You could use `pd.crosstab` or `pd.pivot_table` to generate the table.

```python
# YOUR CODE HERE
# write a comment how to approach top 20 beers
# top 20 beers: most common reviewed beer, count for beer/name
top20_beer=beer_reviews['beer/name'].value_counts().head(20).index
top20_beer
```

```python
#df.drop(df.columns[1], axis=1, inplace = True)

# top20_beerdf=top20_beer.to_frame()
# top20_beerdf=top20_beerdf.drop(top20_beerdf.columns[0],axis=1)
# top20_beerdf
```

```python
beer_filtered=beer_reviews[beer_reviews['beer/name'].isin(top20_beer)]
beer_filtered
```

```python
pv_beer_reviews=pd.pivot_table(beer_filtered, index='beer/name', columns='goodness', aggfunc='size')
pv_beer_reviews = pv_beer_reviews.div(pv_beer_reviews.sum(axis=1), axis=0)
pv_beer_reviews
```
